```python
import numpy as np
import matplotlib.pyplot as plt
```

**Exploring the Activation Sigmoid Function**

In this section, we dive into the mathematical formulation and graphical representation of the sigmoid function, a popular activation function in neural networks. The sigmoid function is defined as:

$$
S(x) = \frac{1}{1 + e^{-x}}
$$

The sigmoid function is a mathematical formula that maps any given input to an output between 0 and 1. In the context of neural networks, it's used as an activation function to introduce non-linearity into the model, allowing the network to learn complex patterns and make better predictions. The function uses the exponential function to squash values, making it especially useful for binary classification problems in machine learning.

We use the `numpy` library to create a set of values for $ x $ from -5 to 5. This helps us see the typical `S` shape of the sigmoid curve clearly As the value of $ x $ gets larger, the result of $ S(x) $ gets closer to 1. When $ x $ gets smaller, $ S(x) $ approaches 0. This means that the sigmoid function always gives a result between 0 and 1, which is why it's so useful in machine learning for tasks like deciding between two classes.

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
    
x = np.linspace(-5, 5, 400)
y = sigmoid(x)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="Sigmoid Function")
plt.title("Sigmoid Function")
plt.xlabel("x")
plt.ylabel("S(x)")
plt.legend()
plt.grid(True)

# Adding a text description
description = r"$S(x) = \frac{1}{1 + e^{-x}}$"
plt.text(-4.5, -0.2, description, fontsize=14, bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'))

plt.show()
```
