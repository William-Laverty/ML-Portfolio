# **Exploring the Sigmoid Function**

In this section, we dive into the mathematical formulation and graphical representation of the sigmoid function, a popular activation function in neural networks. The sigmoid function is defined as:

$$
S(x) = \frac{1}{1 + e^{-x}}
$$

The `matplotlib` library is then employed to plot these values, allowing us to observe the sigmoid function's behavior. As \(x\) approaches positive infinity, \(S(x)\) nears 1, and as \(x\) approaches negative infinity, \(S(x)\) nears 0.
